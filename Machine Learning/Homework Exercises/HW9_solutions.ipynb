```python
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
```

## Exercise 5 - Support Vector Machines

```python
dataset_a = sio.loadmat('HW9a.mat')

train_x = dataset_a['train_X']
train_y = dataset_a['train_y']
test_x = dataset_a['test_X']
test_y = dataset_a['test_y']
```

```python
pos_train_x_cl1 = np.where(train_y == 1)[0]
pos_train_x_cl2 = np.where(train_y == -1)[0]
```

```python
plt.scatter(train_x[pos_train_x_cl2,0],train_x[pos_train_x_cl2,1])
plt.scatter(train_x[pos_train_x_cl1,0],train_x[pos_train_x_cl1,1])
```

#### a.) linear kernel

```python
from sklearn.svm import SVC as svm

clf = svm(gamma='scale', kernel='linear')
clf.fit(train_x, np.ravel(train_y)) 
preds = clf.predict(test_x)
```

```python
p_error = np.sum(preds[...,None] != test_y)/len(preds); p_error
```

#### b.) polynomial kernel 

```python
clf = svm(gamma='scale', kernel='poly', degree=4)
clf.fit(train_x, np.ravel(train_y)) 
preds = clf.predict(test_x)
```

```python
p_error = np.sum(preds[...,None] != test_y)/len(preds); p_error
```

```python
plot_decision_regions(train_x, np.ravel(train_y), clf, legend=2)
```

```python
clf = svm(gamma='scale', kernel='poly', degree=10)
clf.fit(train_x, np.ravel(train_y)) 
preds = clf.predict(test_x)
```

```python
p_error = np.sum(preds[...,None] != test_y)/len(preds); p_error
```

```python
plot_decision_regions(train_x, np.ravel(train_y), clf, legend=2)
```

#### c.) rbf kernel

```python
clf = svm(gamma='scale', kernel='rbf')
clf.fit(train_x, np.ravel(train_y)) 
preds = clf.predict(test_x)
```

```python
p_error = np.sum(preds[...,None] != test_y)/len(preds); p_error
```

```python
plot_decision_regions(train_x, np.ravel(train_y), clf, legend=2)
```

## Exercise 6 - Neural nets

```python
dataset_b = sio.loadmat('HW9b.mat')

train_x = dataset_b['train_X']
train_y = dataset_b['train_y']
test_x = dataset_b['test_X']
test_y = dataset_b['test_y']
```

```python
pos_train_x_cl1 = np.where(train_y == 1)[0]
pos_train_x_cl2 = np.where(train_y == -1)[0]
```

```python
plt.scatter(train_x[pos_train_x_cl1,0],train_x[pos_train_x_cl1,1])
plt.scatter(train_x[pos_train_x_cl2,0],train_x[pos_train_x_cl2,1])
```

```python
from sklearn.neural_network import MLPClassifier

clf = MLPClassifier(solver='lbfgs', hidden_layer_sizes=3, random_state=1, activation='tanh') 
clf.fit(train_x, np.ravel(train_y))
preds = clf.predict(test_x)
```

```python
p_error = np.sum(preds[...,None] != test_y)/len(preds); p_error
```

```python
plot_decision_regions(train_x, np.ravel(train_y), clf, legend=2)
```
